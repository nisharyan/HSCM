```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flood = pd.read_excel("Flood Data.xlsx")
```

```python
flood
```

```python
flood = flood.set_index('Year')
```

```python
fig, ax = plt.subplots(figsize=(20,10))  
flood_heatmap = sns.heatmap(flood, linewidths=0.5)
plt.title("Frequency of occurence of flood related news over the years 2001-2022")
plt.xlabel("Months")
# plt.show() 
plt.savefig('flood Heatmap.png', transparent=True)
```

```python
cholera = pd.read_excel('Cholera data.xlsx')
```

```python
cholera = cholera.set_index('Year')
```

```python
fig, ax = plt.subplots(figsize=(20,10))  
flood_heatmap = sns.heatmap(cholera, linewidths=0.5)
plt.title("Frequency of occurence of water borne diseases related news over the years 2001-2022")
plt.xlabel("Months")
plt.show() 
# plt.savefig('flood Heatmap.png', transparent=True)
```
